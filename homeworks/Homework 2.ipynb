## Rules for Coding Submissions

As we progress through the semester, we will add more programming standards. For now, here are a few you can reasonable be expected to follow on this assignment at this point of the semester:

1 - Your code must be easy to read. Bunched together, difficult to read code will receive point deductions.
<ul><li>If you are unsure, ask a TA. The TAs will be the ones reading your code, so if they are able to easily read your code, then you are in good shape.</li></ul>

2 - Your code must contain variables that indicate what you are performing in your code. Imagine if Lord of the Rings had four Hobbits, a, b, c, and d instead of Frodo, Samwise, Merry, and Pippin. That would be a tough read! 
<ul>
    <li>Good example: <code>double radius_solution_1;</code></li>
    <li>Bad example: <code>double x;</code></li>
</ul>

3 - Your code must have proper tabbing to show levels (i.e., if statements, for loops, ect.)

<b>Here is a good example:</b>

    #include <stdio.h>
    #include <stdlib.h>

    int main(){
         /* Input variables */
        int input_1 = 5;
        int input_2 = 6;

        /* Sum operation */
        int sum = input_1 + input_2;
        fprintf(stdout,"The sum of %d and %d is %d\n", input_1, input_2, sum);
        return 0;
    }


<b>Here is a bad example of code that is often submitted by novice programmers:</b>

    #include <stdio.h>
    #include <stdlib.h>
    int main(){
    int x=5,y=6;
    int z=x+y;
    fprintf(stdout,"%d\n", z);
    return 0;
    }

## Problem 4

In <code>Prob4.c</code>, perform the following task:

> Design, write, compile, and run a C program that calculates and prints the value of the current flowing through an RC circuit. The circuit consists of a battery connected in series to a switch, a resistor, and a capacitor. When the switch is closed, the current I flowing through the circuit is given by the formula:

$$ { i = { E \over R } { e ^ { -t \over {RC} } } } $$

<ul>
    <li>E is the voltage of the battery in volts</li>
    <li>R is the value of the resistor in ohms</li>
    <li>C is the value of the capacitor in farads.</li>
    <li>t is the time in seconds after the switch is closed</li>
    <li>e is Euler's number, which is 2.71828 (rounded to five decimal places)</li>
</ul>

Your program should complete the following tasks:

1 - First time through, get and print the result (to 5 decimal places) for when Voltage is 20 volts, R = 10 ohms, C = 0.00440 farads, and t = 0.02300 seconds
> Hint, when printing a double, use <code>%.5lf</code> (long float for <code>l</code>, not the number <code>1</code>) as the output specifier to get 5 decimal places

2 - Second time through, get and print the result (to 5 decimal places) when Voltage is 35 volts, R = 10 ohms, C = 0.016 farads, and t = 0.067 seconds

> You may put in double variables to pow to simplify this code, as you did in Prob3.<br>
> You must set all the numbers as variables. You may not put them into an equation as numbers

For example, this line of code is not permitted: 

    fprintf( stdout, "%lf\n", 20/10 * 2.71828 * pow( -1 * 0.067, 10 / 0.067 ) );

Print both results to the terminal. Here is an example test run.
