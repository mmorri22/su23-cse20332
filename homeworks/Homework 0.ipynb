## Section 1 - Tabby, OpenSSH, and VPN Introduction

### 1.1 - What is an SSH, and what are Tabby and OpenSSH?
The <b>Secure Shell (SSH)</b> is a communication protocol between a <b>client</b> (you) and a <b>Server</b> (ND Machines). The general overview is shown below. The crucial knowledge you should take away is that SSH permits you to run commands on your laptop, and the program is executed on another machine.

<a href = "https://tabby.sh/">Tabby</a> is a customizable cross-platform terminal app for local shells, serial, SSH and Telnet connections. This program is open source and works well across Windows, MacBook, and Linux operating systems. Tabby is useful for SFTP and Zmodem file transfers, X11, and port forwarding.

<ul>
<li>One benefit of Tabby as opposed to programs liked PuTTy and XTerm is that you can use one program to <b>run several tasks</b>. Additionally, for students who prefer using multiple Vim tabs as opposed to File Transfer Programs (like the FileZilla program I will ask you to install later), you are welcome to use this approach.</li>   
    <li>Another benefit of Tabby is that there are many <a href = "https://tabby.sh/about/features">features</a> which allow you to change the color and size of the Terminal to make observation <b>easier on your eyes</b>. There are also several plugins you can install and play with to enhance your programming experience.</li>
</ul>

### 1.3 - Installing Tabby on a Windows Machine

This section is to assist in helping students with PCs and Windows laptops install the Tabby SSH client on their machine, and be able to log into the Notre Dame servers.
  
1 - Go to the Tabby download page: https://github.com/Eugeny/tabby/releases/tag/v1.0.169
    
Next, scrolls down past the "Assets" and select the appropriate "setup.exe" file
For the Summer 2023 semester, the most up-to-date version is tabby-1.0.169-setup.exe.
    
Run the setup.exe file and follow the prompts.
    
> Note: There is a video after Step 3 where I show how to perform Steps 2 and 3.

2 - Once installed, type "Tabby Terminal" in the Windows search bar. This command will bring up the Tabby Terminal app for you to run.

### 1.5 - Installing OpenSSH on an Ubuntu OS
      
1 - Install <b>OpenSSH</b> on Ubuntu to be able to access the ND server
<ul>
    <li>Open the terminal application for Ubuntu desktop</li>
    <li>Type the following command: <code>sudo apt-get install openssh-server</code></li>
    <b>Note</b>: You will need root access, which you should have on your own laptop.
    <li>Enable the ssh service by typing the following command: <code>sudo systemctl enable ssh</code></li>
    <li>Start the ssh service by typing the following command: <code>sudo systemctl start ssh</code></li>
</ul>

2- From the Terminal, enter the command: <code>ssh -Y [Your_NetID]@student00.cse.nd.edu</code>
