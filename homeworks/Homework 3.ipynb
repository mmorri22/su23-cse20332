### How Insertion Sort Works

> <b>Note:</b> Below the description of insertion sort, there is a Visualize It! where I show the algorithm in action.

We select a location and then move the data at that location to the left until it is no longer smaller than the data at the previous array location.

<ul>
    <li>We split the array into a sorted part and an unsorted part.</li>
    <ul><li>So initially, the value is unsorted, so you begin at array element 0.</li></ul><p></p>
    <li>The sorted part will go from the beginning of the array to some index, and the rest of the array will form the unsorted part.</li><p></p>
    <li>We start by saying that only the first element forms the sorted part.</li><p></p>
    <li>In each step, we take away one element of the unsorted portion of the array and insert it in its place in the sorted part of the array.</li> 
    <ul><li>This increases the length of the sorted part of the array by one.</li></ul><p></p>
    <li>We are done after n - 1 steps because we increase the length of the sorted part of the array by 1</li><p></p>
    <li>Once the length of the sorted part of the array is n, we know that the array is sorted.</li><p></p>
</ul>

### Example Sort

Consider the integer array: {22 17 10 84 99 7 36 8 15}. The sort will go like this:

<code>22 17 10 84 99 7 36 8 15</code>

Start at 17, move it left until it is no longer smaller.

<code>17 22 10 84 99 7 36 8 15</code>

Start at 10, move it left until it is no longer smaller.

<code>10 17 22 84 99 7 36 8 15</code>

84 cannot be moved.
99 cannot be moved.

Start at 7, move it left until it is no longer smaller.

<code>7 10 17 22 84 99 36 8 15</code>

Start at 36, move it left until it is no longer smaller.

<code>7 10 17 22 36 84 99 8 15</code>

Start at 8, move it left until it is no longer smaller.

<code>7 8 10 17 22 36 84 99 15</code>

Start at 15, move it left until it is no longer smaller.

<code>7 8 10 15 17 22 36 84 99</code>

### How Binary Search Works

> <b>Note:</b> Below the description of insertion sort, there is a Visualize It! where I show the algorithm in action.

Consider an array with 1,024 elements in it. Even if we sorted the array, there are <i>worst-case</i> scenarios where proving an element is not in the array would require checking every element.
<ul><li><b>Example:</b> If the largest number is 4,000, and we check for 4,001, we would have to go through every element in the array!</b></li></ul>

You've probably thought to yourself: "What if I check the largest and smallest values?" But then you would have to start at the beginning and end, checking both elements. The worst case scenario is now if the element is in the middle! But you are on the right track...<p></p>

### The Binary Search Algorithm

The main idea is that you cut the array in half by testing the midpoint. If the element we are searching for is larger than the midpoint, then there is no need to search any elements to the left. Likewise, if the element we are searching for is smaller than the midpoint, then there is no need to search any elements to the right.

So we can just continuously update our min and max values, check the midpoint, and update again until min and max are equal. If min and max are equal, and the value is not in the array, then the element is not in the array.

Here is the basic algorithm outline:

<ul>
    <li>Set the <code>minimum</code> to 0 and the <code>maximum</code> to LENGTH - 1</li><p></p>
    <li>Continue checking while the <code>minimum</code> is less than or equal to the <code>maximum</code></li><p></p>
        <ul>
            <li>Find the initial <code>midpoint</code> by adding the <code>minimum</code> and <code>maximum</code>, divide the sum by two</li><p></p>
            <li>If the element of the array at the <code>midpoint</code> is equal to the value we are searching for, <code>return</code> the index of <code>midpoint</code></li><p></p>
            <li>If the element of the array at the <code>midpoint</code> is less than to the value we are searching for, set the <code>maximum</code> to one less than the <code>midpoint</code></li><p></p>
            <li>If the element of the array at the <code>midpoint</code> is greater than to the value we are searching for, set the <code>minimum</code> to one more than the <code>midpoint</code></li><p></p>
        </ul>
    <li>If we get here, then the element is not in the array. <code>return -1</code></li><p></p>
</ul>
